```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder

print("Libraries Imported Successfully")

data = pd.read_csv('/content/sample_data/iris.csv', compression='zip')
print(data.head())
```

```python
print(data.info())
print('\nshape of dataset:', data.shape)
print('\nStatical Summary')
print(data.describe())
```

```python
print(data.isnull().sum)
```

```python
data_dup = pd.concat([data, data.iloc[:5]], ignore_index=True)

print("Shape before removing duplicates:", data_dup.shape)

data_dup.drop_duplicates(inplace=True)

print("Shape after removing duplicates:", data_dup.shape)
```

```python
encoder = LabelEncoder()
data['Species'] = encoder.fit_transform(data['Species'])
data.head()
```

```python
plt.figure(figsize=(6,4))
plt.plot(data['SepalLengthCm'])
plt.plot(data['SepalLengthCm'])
plt.xlabel('index')
plt.ylabel('sepal length')
plt.title('sepal length')
plt.show()
```
